```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_image=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\matn.png")
gray_image=cv2.cvtColor(input_image,cv2.COLOR_BGR2GRAY)
_,binary_image=cv2.threshold(gray_image,127,255,cv2.THRESH_BINARY_INV)
plt.figure(figsize=[12,8])
plt.subplot(131);plt.imshow(input_image,cmap="gray"),plt.title("orginal")
plt.subplot(132);plt.imshow(gray_image,cmap="gray"),plt.title("gray_image")
plt.subplot(133);plt.imshow(binary_image,cmap="gray"),plt.title("binary_image")
```

```python
kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(15,10))
dilation=cv2.dilate(binary_image,kernel,iterations=2)
outputimg=input_image.copy()
contours,hierarchy=cv2.findContours(dilation,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
all_contours=cv2.drawContours(input_image,contours,-1,(0,255,0),4)
plt.figure(figsize=[15,8])
plt.subplot(121);plt.imshow(dilation,cmap="gray"),plt.title("after dilation")
plt.subplot(122);plt.imshow(all_contours,cmap="gray"),plt.title(" with nall_contours")
```

```python
MIN_CONTOUR_AREA=5
for contour in contours:
    if cv2.contourArea(contour) > MIN_CONTOUR_AREA:
        [intX,intY,intW,intH]=cv2.boundingRect(contour)
        cv2.rectangle(outputimg,(intX,intY),(intX+intW,intY+intH),(0,0,255),2)
        imgROI=binary_image[intY:intY+intH,intX:intX+intW]
        subContours,subHierarchy=cv2.findContours(imgROI.copy(),
                                                 cv2.RETR_EXTERNAL,
                                                 cv2.CHAIN_APPROX_SIMPLE)
        for subContour in subContours:
            [pointX,pointY,width,height]=cv2.boundingRect(subContour)
            cv2.rectangle(outputimg,(intX+pointX,intY+pointY),(intX+pointX+width,intY+pointY+height),(255,0,0),2)
plt.figure(figsize=[15,8])
plt.imshow(outputimg[...,::-1])
        
```

```python
image=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\x.jpg")
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray_blur=cv2.GaussianBlur(gray,(31,31),0)
thresh=cv2.adaptiveThreshold(gray_blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,11,1)
kernel=np.ones((5,5),np.uint8)
closing=cv2.morphologyEx(thresh,cv2.MORPH_CLOSE,kernel)
contours,hierarchy=cv2.findContours(closing,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
drawing_image=image.copy()
for cnt in contours:
    if len(cnt)<30:
        continue
    ellipse=cv2.fitEllipse(cnt)
    cv2.ellipse(drawing_image,ellipse,(0,255,0),2)
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Orginal")
plt.subplot(132);plt.imshow(closing,cmap="gray"),plt.title("closing")
plt.subplot(133);plt.imshow(drawing_image[...,::-1]),plt.title("drawing")
```

```python
image=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\x.jpg")
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
gray_blur=cv2.GaussianBlur(gray,(31,31),0)
thresh=cv2.adaptiveThreshold(gray_blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,11,1)
kernel=np.ones((5,5),np.uint8)
closing=cv2.morphologyEx(thresh,cv2.MORPH_CLOSE,kernel)
contours,hierarchy=cv2.findContours(closing,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
drawing_image=image.copy()
for cnt in contours:
    area=cv2.contourArea(cnt)
    if area < 5000 or area > 50000:
        continue
    if len(cnt)< 5:
        continue
    ellipse=cv2.fitEllipse(cnt)
    cv2.ellipse(drawing_image,ellipse,(0,255,0),2)
plt.figure(figsize=[15,15])
plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Orginal")
plt.subplot(132);plt.imshow(closing,cmap="gray"),plt.title("closing")
plt.subplot(133);plt.imshow(drawing_image[...,::-1]),plt.title("drawing")
```
